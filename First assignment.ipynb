<b>Question 1:<header>Type Conversion</header></b> 
<p> convert the string "42" to an integer and print the result.</p>

```python
var= "42"
int_var= int(var)
print(f"Converted {var} ({type(var)}) to {int_var}({type(int_var)})")
```

<b>Question 2</b> 
<p> Use string indexing to extract the third character from the string "Python".</p>

```python
aa= "python"
count=0
for ch in aa:
    count+=1
    if count==3:
        print(ch)
```

<b>Question 3</b> 
<p> Take a sentence as input and print only the first five words.</p>

```python
bb = input("Enter a sentence: ")
words = bb.split()
for i in range(5):
    print(words[i])
```
