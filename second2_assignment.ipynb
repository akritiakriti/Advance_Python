- Use a `try-except` block to handle `ZeroDivisionError`.
- If a division by zero occurs, return "Cannot divide by zero" instead of crashing.

````python
def safe_divide(a,b):
    try:
        c= a/b
        print(f"{a} / {b} = {c}")
        
    except ZeroDivisionError as e:
        print("Cannot divide by zero.")
    
    finally:
        print("Thenks!!")
        
safe_divide(4,2)
safe_divide(4,0)

    
````

## Question 16: OOP - Encapsulation
Create a class `BankAccount` with a private attribute `_balance`.
- Implement methods `deposit(amount)` and `withdraw(amount)` to modify the balance.
- Ensure that `withdraw` does not allow taking out more money than the current balance.
- Create a `get_balance()` method to safely access the private balance.

````python
class BankAccount:
    def __init__(self, balance_initial=0):
        self._balance = balance_initial
        
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Amount must be positive!")
            
    def withdraw(self, amount):
        if amount <= 0:
            print("Amount must be positive!")
        elif amount > self._balance:
            print("Insufficient balance!")
        else:
            self._balance -= amount
            print(f"Withdrawn: {amount}")

    def get_balance(self):
        return self._balance
        
b = BankAccount(10_000)
print("Balance:", b.get_balance())

b.deposit(4000)
b.withdraw(15_000)
b.withdraw(5000)
print("Balance:", b.get_balance())


````

## Question 19: Variable Arguments (*args and **kwargs)
Write a function `student_report` that accepts:
- A mandatory argument `name`.
- Arbitrary positional arguments (`*args`) for subject scores.
- Arbitrary keyword arguments (`**kwargs`) for additional details (like `city`, `age`).
- The function should print the name, calculate the average of the scores, and print the additional details.

#### Explanation:
`*args` – Arbitrary Positional Arguments  
- Purpose: To pass any number of positional arguments to a function.  
- `*args` collects them as a tuple.  
- Inside a function, `*args` acts as a tuple.  

`**kwargs` – Arbitrary Keyword Arguments  
- Purpose: To pass any number of keyword arguments (name=value) to a function.  
- `**kwargs` collects them as a dictionary.  
- Keyword arguments are passed as `key=value` pairs.  
- Inside the function, `**kwargs` is a dictionary.


````python
def student_report(name, *args, **kwargs):
    print(f"Name: {name}")

    #avg
    if args:
        avg = sum(args) / len(args)
        print(f"Average_score: {avg}")
    else:
        print("No arguments provided!")

    #additional details
    if kwargs:
        for key,value in kwargs.items():
            print(f"{key}: {value}")
            
student_report(
    "Akriti",
    85, 90, 78, 92,       # positional arguments -> subject scores
    city="Kathmandu",
    age=17,
    grade="A"
)
    
````

## Question 20: Comprehensive System
Design a simple "Library Management System" using OOP and Exception Handling.
- Create a class `Library` with a list of available books.
- Add methods to `borrow_book(book_name)` and `return_book(book_name)`.
- If a user tries to borrow a book that isn't in the list, raise a custom exception `BookNotFoundError` (or use a standard `ValueError`).
- Ensure the state of the library updates correctly after each transaction.

````python
class BookNotFoundError(Exception):
    pass
    
class Library:
    def __init__(self, book):
        self.book= book

    def borrow_book(self,book_name):
        if book_name in self.book:
            self.book.remove(book_name)
            print(f"You've borrowed {book_name}")
        else:
            raise BookNotFoundError(f"{book_name} not available!")

    def return_book(self,book_name):
        self.book.append(book_name)
        print(f"Thankyou for returning {book_name}.")

    def book_show(self):
        if self.book:
            print("Available books:")
            for book in self.book:
                print(f" - {book}")
        else:
            print("No books available!")
            
library = Library(["Python Basics", "Data Science 101", "AI for Beginners"])
library.book_show()

#borrow a book
try:
    library.borrow_book("Python Basics")
except BookNotFoundError as e:
    print(e)

#Borrow a book that isn't available.
try:
    library.borrow_book("Deep Learning")
except BookNotFoundError as e:
    print(e)

#Return a book
library.return_book("Python Basics")

library.book_show()
````
