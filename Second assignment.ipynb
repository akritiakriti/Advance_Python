## Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.


````python
marks = [45, -3, 88, "NA", 102, None, 67]

def clean(marks):
    #empty list
    cleaned=[]
    invalid= 0

    for m in marks:
        if isinstance (m,(int,float)) and 0<=m<=100:
            cleaned.append(m)

        else:
            invalid+=1
    total= len(marks)
    Percentage=  (len(cleaned) / total) * 100
    return cleaned, invalid,Percentage
a,b,c= clean(marks)

print(f"Cleaned list: {a}\nNumber of invalid entries: {b}\nPercentage of valid marks: {c}")

    
````

## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.


````python
           
print("The prime numbers between 1 and 50:")

def is_prime():
    prime = []
    for num in range(2, 50):
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break  
        else:
            prime.append(num)  # else of for → no divisors found
    return prime

print(is_prime())

````

## Question 9: Function Composition
Write functions to:
1. Calculate total marks  
2. Calculate average  
3. Classify result (Distinction / Pass / Fail)

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.


````python
size = 5
marks = []

for i in range(size):
    m = int(input(f"Enter mark {i+1}: "))
    marks.append(m)

def student_result(marks):
    def total():
        sum_marks = 0
        for mark in marks:
            sum_marks += mark
        return sum_marks

    def avg():
        return total() / len(marks)
    
    def result():
        average = avg()
        print(f"Total marks: {total()}")
        print(f"Average marks: {average:.2f}")
        if average >= 80:
            print("Grade: Distinction")
        elif average >= 60:
            print("Grade: Good")
        elif average >= 40:
            print("Grade: Try more")
        else:
            print("Grade: Fail")
    
    result() #internal call to trigger above 2 functions

student_result(marks)
````

<b>Modular design</b> means breaking a program into smaller, independent parts.( modules or functions).Each module has specific task.Its beneficial as it increases:
- readability
- Maintainability
- Resuability
- Scalability


## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python
def get_info(raw):
    try:
        #split main parts
        parts= raw.split("|")
        if len(parts) != 4:
            return "Invalid Format"

        name= parts[0]
        roll=int(parts[1])
        age=int(parts[2])
#Validate age
        if age <= 16:
            return "Invalid age"

#Split and convert marks
        marks_str= parts[3].split(",")
        marks=[]

        for m in marks_str:
            mark= int(m)
            if mark < 0 or mark >100:
                return "Invalid marks"
            marks.append(mark)

#return dict
        return{
            roll:{
                "name": name,
                "age": age,
                "marks":marks
                }
        }
    except:
#handle unexpected error
        return "Invalid Input!!"
        
#input
raw_input= input("\nEnter the data:\n")
result= get_info(raw_input)
print(f"\n{result}")
````

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
data = [10, "20", 30.5, "abc", None, True]

def seperation(raw):
#empty list for numeric and non-numeric
    numeric = []
    non_numeric = []

    for item in raw:
        try:
            num=float(item)
            numeric.append(num)
        except:
    #if conversion fails, its not a numeric value
            non_numeric.append(item)
    return numeric, non_numeric

    
num,non_num= seperation(data)
print("\nNumeric values: ",num)
print("\nNumeric values: ",non_num)
````

`True behaves like a number` because in python bool is a subclass of int. `True` behaves like a `1` and False as 0. So, `float(True)->1.0`

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

def analyze(records):
    results= {}
    for student,marks in records.items():
        avg= sum(marks)/len(marks)
        results[student]= {'average': avg}

    averages= []
    for student in results:
        averages.append(results[student]["average"])

    max_avg = max(averages)
    min_avg = min(averages)

    for student in results:
        avg= results[student]["average"]
        if avg== max_avg:
            results[student]["status"]="Top Performer"
        
        elif avg== min_avg:
            results[student]["status"]="Low Performer"
        
        else:
            results[student]["status"]="Average Performer"
            
    return results
final = analyze(records)
print(final)
        
````

## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

def analyze_marks(school_data):
    average={}
    topper={}
    total_marks=0
    total_std=0

    for class_name, students in school_data.items():
        top_student = None
        top_avg= 0

        for student, marks in students.items():
            avg= sum(marks)/len(marks)
            average [student] = avg

            if avg > top_avg:
                top_avg = avg
                top_student = student

            total_marks+=sum(marks)
            total_std+=len(marks)
        topper[class_name] = (top_student, top_avg)

    school_average = total_marks / total_std
    return {
        "class_toppers" : topper,
        "school_average" : school_average,
        "student_average" : average
    }

final_result= analyze_marks(school_data)
print(final_result)
````

<b>Nested Dictionaries</b> are useful for hierarchical data bc:
- Each class can have multiple students
- Easy to access class
- Makes aggregration(like class topper, school average) simple
- Increases readability and matches real world structure
